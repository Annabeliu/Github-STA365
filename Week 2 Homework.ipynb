## 1. Make the "Classic Bayes" figure, but instead of a single likelihood, plot the individual likelihoods of each observation. 

*Hint: Consider the code...*
    
    ```python
    for xi in range(x):
        print(xi)
    ```     

````python
import numpy as np; from scipy import stats; import matplotlib.pyplot as plt

n = 3; x = stats.norm(loc=0,scale=1).rvs(n); 
theta_0,tau = 4,1/2; 
parameter_domain = np.linspace(-10,10,1001); 
prior = stats.norm(loc=theta_0, scale=1/tau**0.5).pdf(parameter_domain) 
c = 30

plt.figure(figsize=(16,5)) 
plt.plot(parameter_domain, prior, label="prior")
for i in range(0,3):
    likelihood = stats.norm.pdf(x[i]*np.ones(parameter_domain.shape), loc=parameter_domain, scale=1)
    joint = c*likelihood*prior # elementwise multiplication 
    posterior = stats.norm(loc=(tau*theta_0+x.sum())/(tau+n), scale=1/(tau+n)**0.5).pdf(parameter_domain)
    plt.plot(parameter_domain, likelihood, label="likelihood "+str(i+1)+"") 
    plt.plot(parameter_domain, posterior, label="posterior")
    plt.plot(parameter_domain, joint, linestyle="--", label="joint "+str(i+1)+"  (times scalar "+str(c)+")")

plt.legend(); 
plt.title("Visualize various priors, likelihoods, posteriors... and joint distribution (which is proportional to...)"); 
````

## 2. Make the "Classic Bayes" figure for the normal-normal model on a log scale. 

*Hint: plot loglikehood, prior logpdf, and the log of the joint distribution using* `stats.norm.logpdf()`

````python
import numpy as np; from scipy import stats; import matplotlib.pyplot as plt

n = 3; x = stats.norm(loc=0,scale=1).rvs(n); 
theta_0,tau = 4,1/2; 
parameter_domain = np.linspace(-10,10,1001); 
prior = stats.norm(loc=theta_0, scale=1/tau**0.5).logpdf(parameter_domain) 
c = 1/50

plt.figure(figsize=(16,5)) 
plt.plot(parameter_domain, prior, label="prior")
for i in range(0,3):
    likelihood = stats.norm.logpdf(x[i]*np.ones(parameter_domain.shape), loc=parameter_domain, scale=1)
    joint = c*likelihood*prior # elementwise multiplication 
    posterior = stats.norm(loc=(tau*theta_0+x.sum())/(tau+n), scale=1/(tau+n)**0.5).logpdf(parameter_domain)
    plt.plot(parameter_domain, likelihood, label="likelihood "+str(i+1)+"") 
    plt.plot(parameter_domain, posterior, label="posterior")
    plt.plot(parameter_domain, joint, linestyle="--", label="joint "+str(i+1)+"  (times scalar "+str(c)+")")

plt.legend(); 
plt.title("Visualize various priors, likelihoods, posteriors... and joint distribution (which is proportional to...)"); 
````

# 4. How do ***log likelihoods*** and ***posteriors*** differ from each other as a result of ***improper priors*** discussed below and ***marginal likelihood constants***? 



What's the difference between a likelihood and posterior when improper prior is used? 

Hint: is has something to do with the marginal likelihood constant.

The improper posterior is proportional to the product of the likelihood and the improper prior, without a proper normalizing constant.

When an improper prior is used, the marginal likelihood, which is the integral of the likelihood times the prior over all possible parameter values, may not exist or may be infinite. The constant of proportionality in the posterior is determined by this marginal likelihood.

The key difference is that, in the presence of an improper prior, the posterior may not be a proper probability distribution because the marginal likelihood is not well-defined.
